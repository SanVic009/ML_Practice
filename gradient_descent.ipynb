```python
import numpy as np
from sklearn.datasets import make_regression
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
x, y= make_regression(n_samples= 4, n_features= 1, n_informative= 1, n_targets= 1, noise= 80, random_state= 13)

```

```python
plt.scatter(x, y)
```

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
```

```python
lr.fit(x, y)
print(lr.intercept_)
print(lr.coef_)
```

```python
plt.scatter(x, y)
plt.plot(x, lr.predict(x))
```

```python
y_pred = ((78.35 * x) + 0).reshape(4)
y_pred
```

```python
plt.scatter(x, y)
plt.plot(x, lr.predict(x), color='black', label= 'OLS')
plt.plot(x, y_pred , label = 'b = 0')
plt.legend()
plt.show()
```

```python
m = lr.coef_[0]
b= 0
print(b, m)

loss_slope = -2 * np.sum(y -m*x.ravel() - b)
print(loss_slope)
```

```python
# taking learning rate as 0.1

l = 0.1
step_size = loss_slope*l
step_size
```

```python
#calculating the new intercept

b = b- step_size
b
```

```python
y_pred1 = ((78.35 * x) + b).reshape(4)

plt.scatter(x, y)
plt.plot(x, lr.predict(x), color='black', label= 'OLS')
plt.plot(x, y_pred1, color = 'green', label = 'b = {}'.format(b))
plt.plot(x, y_pred, color= 'red', label= 'b = 0')
plt.legend()
plt.show()
```

#### as we iterate through, we get closer and closer to the regression line
